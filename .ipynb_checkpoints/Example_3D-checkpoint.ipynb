# 3-D Visualization with Cluster Decoder

```python
from qiskit.ignis.verification.topological_codes import RepetitionCode
from qiskit.ignis.verification.topological_codes import GraphDecoder
```

```python
code = RepetitionCode(29,1)
```

```python
code.circuit['0'].draw('mpl')
```

```python
code.circuit['1'].draw('mpl')
```

```python
from qiskit import Aer, execute
raw_results = {}
for log in ['0' , '1']:
    job = execute(code.circuit[log],Aer.get_backend('qasm_simulator'))
    raw_results[log] = job.result().get_counts()
print(raw_results)
```

```python
code.process_results(raw_results)
```

```python
Graph = GraphDecoder(code)
```

## The result used below was from actual quantum computer and let's see it's decoding.

```python
Error_Graph = Graph.make_error_graph('1 0  0000000100010000000000000000 1010000010010000000000000111')['0']
```

## 3-D interactive output is visible in a separate window

```python
Graph.draw_3d_error_graph(Error_Graph)
```

```python
[logical, Edgelist, nodelist]=Graph.cluster_decoding('1 0  0000000100010000000000000000 1010000010010000000000000111')
```

```python
Graph.draw_3d_decoded_graph(Error_Graph,Edgelist,nodelist)
```
